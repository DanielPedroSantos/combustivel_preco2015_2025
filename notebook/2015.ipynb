```python
# Importando as bibliotecas

import pandas as pd 
import numpy as np
import os
from datetime import datetime, date
from google.cloud import bigquery
from dotenv import load_dotenv
import pyarrow


# Carregando variáveis de ambiente
load_dotenv()
```

```python
credencial = os.getenv("GOOGLE_APPLICATION_CREDENTIALS")
project_id = os.getenv("PROJECT_ID")
table_stg = os.getenv("STG_2015")
```

```python
# Inicializa o cliente do BigQuery usando credenciais de serviço
# O parâmetro 'project_id' especifica o projeto GCP onde as queries serão executadas
client = bigquery.Client.from_service_account_json(credencial, project=project_id)


query = f"""
SELECT *
FROM `{table_stg}`
"""
```

```python
# Executa e converte pra DataFrame
resultado = client.query(query)
df = resultado.to_dataframe()
```

```python
df.head()
```

```python
df_2015_silver = df
```

```python

# Substitui espaço e hífen por underline
df_2015_silver.columns = (df_2015_silver.columns
                          .str.lower()
                          .str.replace('[ -]', '_', regex=True)
                          .str.replace('_+', '_', regex=True))
```

```python


# Remove underlines duplicados
df_2015_silver.columns = df_2015_silver.columns.str.replace('_+', '_', regex=True)
```
