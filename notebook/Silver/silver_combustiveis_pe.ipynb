```python
# Importando as bibliotecas

import pandas as pd 
import numpy as np
import os
from datetime import datetime, date
from google.cloud import bigquery
from dotenv import load_dotenv
import pyarrow
import unicodedata



# Carregando variáveis de ambiente
load_dotenv()
```

```python
credencial = os.getenv("GOOGLE_APPLICATION_CREDENTIALS")
project_id = os.getenv("PROJECT_ID")
table_bronze2015 = os.getenv("BRONZE_2015")
table_bronze2016 = os.getenv("BRONZE_2016")
table_bronze2017 = os.getenv("BRONZE_2017")
table_bronze2018 = os.getenv("BRONZE_2018")
table_bronze2019 = os.getenv("BRONZE_2019")
table_bronze2020 = os.getenv("BRONZE_2020")
table_bronze2021 = os.getenv("BRONZE_2021")
table_bronze2022 = os.getenv("BRONZE_2022")
table_bronze2023 = os.getenv("BRONZE_2023")
table_bronze2024 = os.getenv("BRONZE_2024")
table_bronze2025 = os.getenv("BRONZE_2025")
```

```python
# Inicializa o cliente
client = bigquery.Client.from_service_account_json(credencial, project=project_id)

# Mapeia nome da tabela → nome do DataFrame
tabelas = {
    "df_2015": table_bronze2015,
    "df_2016": table_bronze2016,
    "df_2017": table_bronze2017,
    "df_2018": table_bronze2018,
    "df_2019": table_bronze2019,
    "df_2020": table_bronze2020,
    "df_2021": table_bronze2021,
    "df_2022": table_bronze2022,
    "df_2023": table_bronze2023,
    "df_2024": table_bronze2024,
    "df_2025": table_bronze2025,
}
```

```python
# Carrega todas as tabelas
dataframes = {}
for nome, table_stg in tabelas.items():
    print(f"Carregando {nome}...")
    query = f"SELECT * FROM `{table_stg}`"
    dataframes[nome] = client.query(query).to_dataframe()

print("✅ Todos os DataFrames carregados!")
```

```python
# Concatenar todos os DataFrames em um único DataFrame
df_final = pd.concat(dataframes.values(), ignore_index=True)
```

```python
#  Removendo os - e espaços dos nomes das colunas, e convertendo para minúsculas
df_final.columns = (df_final.columns
                          .str.lower()
                          .str.replace('[ -]', '_', regex=True)
                          .str.replace('_+', '_', regex=True))
```

```python
df_final.head()
```

```python
# Função para remover acentos
def remove_acentos(texto):
    if isinstance(texto, str):
        return unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('ASCII')
    return texto
```

```python
# Normaliza CNPJ: remove tudo que não for dígito
df_final["cnpj_da_revenda"] = df_final["cnpj_da_revenda"].astype(str).str.replace(r'\D', '', regex=True)
```

```python
# Remove pontos e espaços extras no nome
df_final["revenda"] = df_final["revenda"].str.replace(r'\.', '', regex=True).str.strip()
```

```python
# Remove acentos de todas as colunas string
df_final = df_final.apply(lambda col: col.map(remove_acentos) if col.dtype == 'object' else col)
```

```python
# Converte tudo para minúsculas
df_final = df_final.apply(lambda col: col.str.lower() if col.dtype == 'object' else col)
```

```python
# Padronização da bandeira: "Vibra Energia" renomeada para "Vibra" após rebranding da marca
df_final["bandeira"] = df_final["bandeira"].str.replace(
    "vibra energia", "vibra", case=False
)
```

```python
# Configurando credenciais

# Caminho do arquivo de credenciais GCP
credencial = os.getenv("GOOGLE_APPLICATION_CREDENTIALS")

# Identificador do projeto no GCP
project_id = os.getenv("PROJECT_ID")

# Tabela Silver no BigQuery
table_id = os.getenv("SILVER_TABLE")

# Inicialização do cliente BigQuery

# Define a variavel de ambiente
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = credencial

# Cria o cliente Bigquery ao projeto configurado
client = bigquery.Client(project=project_id)
```

```python
# Criação e carga da tabela Silver

# Configuração do job de carga no BigQuery
# WRITE_TRUNCATE garante que a tabela Silver seja sempre sobrescrita,
# mantendo apenas o retrato mais recente dos dados de origem
job_config = bigquery.LoadJobConfig(
    write_disposition="WRITE_TRUNCATE"
)

# Envia o DataFrame para o BigQuery,
# criando a tabela automaticamente caso ainda não exista
job = client.load_table_from_dataframe(
    df_final,
    table_id,
    job_config=job_config
)

# Aguarda a finalização do job para garantir consistência da carga
job.result()

# Confirmação visual de sucesso da operação
print("✅ Tabela Silver criada e dados carregados com sucesso")
```
